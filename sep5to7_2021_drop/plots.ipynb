### Load libraries

```python
import pandas as pd

%matplotlib inline
```

### read data

```python
df_eth = pd.read_csv("./eth_data.csv")
df_doge = pd.read_csv("./doge_data.csv")

df_eth.set_index('datetime', inplace=True)
df_doge.set_index('datetime', inplace=True)

df_eth.shape, df_doge.shape 
```

```python
df_eth.head()
```

# Let's take a look at ETH

### ETH - Calculate moving average and plot

```python
col_ma = 'reddit_hype_volatility_index'
df_eth[col_ma+"_ma"] = df_eth[col_ma].rolling(12, min_periods=1).mean()
axes = df_eth[['open',
               col_ma+"_ma"]].plot(figsize=(15,10), 
                                   title='ETH: Open price & Reddit hype volatility index: Cryptoreview.ai',
                                          secondary_y=col_ma+"_ma")

axes.set_ylabel("Open price")
```

```python
df_eth.columns.tolist()
```

```python
cols = [
    'low',
    'high',
#     'open',
#     'close',
#     'volume',
    'reddit_hype_volatility_index_ma',
    'reddit_num_comments',
    'reddit_num_posts',
    'reddit_post_ups',
    ]

axes = df_eth[df_eth.index>='2021-09-05'][cols].plot(figsize=(15,25), subplots=True)
axes[0].set_ylabel('price')
axes[1].set_ylabel('price')
axes[2].set_ylabel('index')
axes[3].set_ylabel('count')
axes[4].set_ylabel('count')
axes[5].set_ylabel('count')
```
